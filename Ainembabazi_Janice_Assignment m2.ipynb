```python
from __future__ import print_function
import numpy as np
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers.core import Dense, Activation
from keras.optimizers import SGD
from keras.utils import np_utils
np.random.seed(16771) # reproducibility
# network and training
NB_EPOCH = 20
BATCH_SIZE = 128
VERBOSE = 1
NB_CLASSES = 10 # number of outputs = number of digits
OPTIMIZER = SGD() # optimizer, explained later in this chapter
N_HIDDEN = 128
VALIDATION_SPLIT=0.2 # how much TRAIN is reserved for VALIDATION
# data: shuffled and split between train and test sets
(X_train, y_train), (X_test, y_test) = mnist.load_data()
#X_train is 60000 rows of 28x28 values --> reshaped in 60000 x 784 
RESHAPED = 784
#
X_train = X_train.reshape(60000, RESHAPED)
X_test = X_test.reshape(10000, RESHAPED)
X_train = X_train.astype('float32')
# normalize
X_train /= 255
X_train /= 255
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
# convert class vectors to binary class matrices
Y_train = np_utils.to_categorical(y_train, NB_CLASSES)
Y_test = np_utils.to_categorical(y_test, NB_CLASSES)
# M_HIDDED hidden layers
# 10 outputs
# final stage is softmax
model = Sequential()
model.add(Dense(N_HIDDEN, input_shape=(RESHAPED,)))
model.add(Activation('relu'))
model.add(Dense(NB_CLASSES))
model.add(Activation('softmax'))
model.summary()
model.compile(loss='categorical_crossentropy',
              optimizer=OPTIMIZER,
              metrics=['accuracy'])
history = model.fit(X_train, Y_train,
                   batch_size=BATCH_SIZE, epochs=NB_EPOCH,
                   verbose=VERBOSE, validation_split=VALIDATION_SPLIT)
score = model.evaluate(X_test, Y_test, verbose=VERBOSE)
print("Test score:", score[0])
print('Test accuracy:', score[1])
```

## Results Summary and Analysis

### Summary of results

- **Experiment 1 (N_HIDDEN = 64):**
    - Test Accuracy: 0.1659
    
- **Experiment 2 (N_HIDDEN = 256):**
   - Test Accuracy: 0.2452 
   
- **Experiment 3 (N_HIDDEN = 512):**
    - Test Accuracy: 0.1812
    
### Analysis of results

The experiments show that changing the number of neurons within the hidden layers ('N_HIDDEN') has a noticeable impact on the models test accuracy:

    1. **More Neurons Helped Initially**:
      - Increasing 'N_HIDDEN' from 64 t0 256 neurons improved the models accuracy from 16.59% to 24.52%. Adding more neurons helped the model learn patterns in the data better, leading to a higher test accuracy.
    
    2. **Too many Neurons led to Overfitting**:
        - When 'N_HIDDEN' was increased to 512, accuracy dropped to 18.12%. With too many neurons, the model became overly complex and focused too much on the training data, causing overfitting and lower performance on new data.
        
    3. **Finding the Right Balance**:
        - Using 256 neurons gave the best accuracy. With 64 neurons, the model underfit (too simple) and with  512 neurons, it overfit (too complex). The 256 neurons provided a balance between learning and generalization.
